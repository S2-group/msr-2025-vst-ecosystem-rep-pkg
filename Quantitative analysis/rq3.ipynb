```python
# !pip install pandas matplotlib numpy
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
df_repositories = pd.read_csv('CSVs Used//final_data_curated - final_data_curated_categorized.csv')
df_repositories['core_functionality'] = df_repositories['core_functionality'].str.strip()

df_users = pd.read_csv('CSVs Used//users_details.csv')
```

```python
# df_comments = pd.DataFrame(data = df_repositories['comments'])
# df_comments['comments_percentages'] = df_repositories['comments'].str.slice(0,2)


df_repositories['cyclomatic_complexity'] = df_repositories['cyclomatic_complexity'].astype(str).apply(lambda x: x.replace(',','.'))
df_repositories['cyclomatic_complexity'] = df_repositories['cyclomatic_complexity'].astype(float)

df_repositories['cognitive_complexity'] = df_repositories['cognitive_complexity'].astype(str).apply(lambda x: x.replace(',','.'))
df_repositories['cognitive_complexity'] = df_repositories['cognitive_complexity'].astype(float)

df_repositories['comments'] = df_repositories['comments'].astype(str).str.slice(0,-1).astype(float)

sns.set()
sns.set_color_codes("pastel")

# sns.set_style("whitegrid")
# sns.set_style("darkgrid", {'axes.grid' : False})
fix, axs = plt.subplots(1, 3, figsize=(10, 5),  dpi=300)
axs[0].set_title('Cyclomatic Complexity')
axs[1].set_title('Cognitive Complexity')
axs[2].set_title('Comment lines density (%)')

sns.violinplot(ax=axs[0], y=df_repositories['cyclomatic_complexity'], color='b', inner='quart', cut=0).set_ylabel("")
sns.violinplot(ax=axs[1], y=df_repositories['cognitive_complexity'], color='b', inner='quart', cut=0).set_ylabel("")
sns.violinplot(ax=axs[2], y=df_repositories['comments'], color='b', inner='quart', cut=0).set_ylabel("")

plt.tight_layout()
plt.savefig('descriptive_cyclo_rq3.png',bbox_inches = 'tight')
```
